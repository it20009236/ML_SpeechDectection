```python
import pandas as pd

# 1. Read CSV file with quotes around the path
df = pd.read_csv("/content/twitter_data.csv")  # Fixed syntax

# 2. Check column names and data
print("First 5 rows:")
print(df.head())

# 3. Check column names
print("\nColumn names:")
print(df.columns.tolist())
```

```python
# Corrected code
df["labels"] = df["class"].map({
    0: "Hate Speech Detected",
    1: "Offensive Language Detected",
    2: "No Hate and Offensive Speech"  # Changed from 3 to 2 (assuming class 2 exists)
})

print(df.head())  # Fixed typo: "primt" → "print"
```

```python
# Corrected code
df = df[['tweet', 'labels']]  # Fixed mismatched quotes and bracket syntax
print(df.head())  # Added print() to display output
```

```python
def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)  # Remove text in brackets
    text = re.sub('https?://\S+|www\.\S+', '', text)  # Remove URLs
    text = re.sub('<.*?>+', '', text)  # Remove HTML tags
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)  # Remove punctuation
    text = re.sub('\n', '', text)  # Remove newlines
    text = re.sub('\w*\d\w*', '', text)  # Remove words with numbers

    # Remove stopwords
    text = [word for word in text.split() if word not in stop_words]
    text = " ".join(text)

    # Stemming
    text = [stemmer.stem(word) for word in text.split()]
    text = " ".join(text)

    return text

df["tweet"] = df["tweet"].apply(clean)
print(df.head())  # Fixed typo in print
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier  # Corrected spelling

# 1. Convert to numpy arrays
x = np.array(df["tweet"])  # Fixed typo: arrwy → array
y = np.array(df["labels"])

# 2. Vectorize text
cv = CountVectorizer()
x = cv.fit_transform(x)  # x is now a sparse matrix

# 3. Split data
X_train, X_test, y_train, y_test = train_test_split(
    x, y,
    test_size=0.33,  # Fixed hyphen (-) to equals (=)
    random_state=42
)

# 4. Initialize and train classifier
clf = DecisionTreeClassifier()  # Corrected spelling
clf.fit(X_train, y_train)
```

```python
# Corrected code
test_data = "I will love you"

# 1. Preprocess the test data using the same cleaning function
cleaned_test_data = clean(test_data)

# 2. Vectorize the cleaned text
test_vector = cv.transform([cleaned_test_data]).toarray()  # Fixed "tramsform" → "transform", "toarrwy" → "toarray"

# 3. Predict and print
print(clf.predict(test_vector))  # Fixed "primt" → "print"
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
plot_tree(clf, feature_names=cv.get_feature_names_out(), class_names=clf.classes_)
plt.show()
```
